```python
def plot_weekend(data,fecha1):
    fig, ax = plt.subplots(figsize=(12,4))

    fecha1 = parse(fecha1) - pd.Timedelta('1D')
    fecha2 = fecha1 + pd.Timedelta('4D')
    ax.set_xlim(fecha1,fecha2)

    ax.scatter(data.index,data.TCAire)
```

```python
plot_weekend(weekend,'2019-06-08')
plot_weekend(weekend,'2019-06-15')
plot_weekend(weekend,'2019-06-22')
plot_weekend(weekend,'2019-07-27')
```
